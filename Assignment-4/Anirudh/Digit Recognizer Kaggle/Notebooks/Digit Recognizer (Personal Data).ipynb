# Making Predictions from Personal Data

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
```

```python
# Importing trained model
from tensorflow.keras.models import load_model

model = load_model('..\Models\TF_cnn_1.2')
```

```python
# Setting Path
path = os.path.join(os.pardir, 'Data', 'Personal_Test_Data')
IMG_SIZE = 28
```

```python
def print_image(img):
    plt.imshow(img, cmap = 'gray')
    plt.show()
```

```python
def print_data():
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
        new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
        print_image(new_array)
```

```python
def get_predictions(model, X):
    y = model.predict(X)
    predictions = []
    for prediction in y:
        predictions.append(np.argmax(prediction))
    return predictions
```

```python
def evaluate_predictions(predictions, data):
    for prediction, img in zip(predictions, data):
        print_image(img)
        print("Prediction :", prediction)
```

```python
data = []

def load_data():
    for img in os.listdir(path):
        try:
            img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
            new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
            data.append(new_array)
        except Exception as e:
            pass
```

```python
print_data()
```

```python
load_data()
```

```python
X = np.array(data).reshape(-1, IMG_SIZE, IMG_SIZE, 1)
X = X / 255
```

```python
predictions = get_predictions(model, X)

evaluate_predictions(predictions, data)
```
